## Handtracking 

```python
import cv2
import mediapipe as mp
import time

capture = cv2.VideoCapture(1)

mpHands = mp.solutions.hands    # class hand
hands = mpHands.Hands() # store to hand

while(True):
    success, img = capture.read()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # convert BGR to RGB (hand class only use RGB encoded)
    results = hands.process(imgRGB) # process the img and store in results


    cv2.imshow("Video", img)
    cv2.waitKey(1)
    # cv2.destroyAllWindows()
```
